```python
from nba_api.stats.endpoints import synergyplaytypes, leaguedashplayerstats
import csv
import json
import pandas as pd
```

```python
play_types= ['Cut', 'Handoff', 'Isolation', 'Misc', 'OffScreen', 'Postup', 'PRBallHandler', 'PRRollman', 'OffRebound', 'Spotup', 'Transition']
years= ['2012-13', '2013-14', '2014-15', '2015-16', '2016-17', '2017-18', '2018-19']
```

```python
for play in play_types:
    for year in years:
        result= synergyplaytypes.SynergyPlayTypes(season= year, player_or_team_abbreviation='P',play_type_nullable=play, type_grouping_nullable='offensive')
        res= result.get_data_frames()[0] 
        filename= "C:/ws/NBASpace/data/statsapi/synergy_{}_{}.csv".format(play, year)
        res.to_csv(filename, index=False)
```

```python
for year in years:
    result= leaguedashplayerstats.LeagueDashPlayerStats(season= year, measure_type_detailed_defense="Advanced")
    res= result.get_data_frames()[0] 
    filename= "C:/ws/NBASpace/data/statsapi/advanced_{}.csv".format(year)
    res.to_csv(filename, index=False)
```
